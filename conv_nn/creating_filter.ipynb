In this task, I create a filter for convolutional neural network, first manually, and then using kernel

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_sample_image
%matplotlib
```

```python
# Load sample image
china = load_sample_image("china.jpg") / 255
flower = load_sample_image("flower.jpg") / 255
images = np.array([china, flower])
batch_size, hight, width, channel = images.shape
```

```python
# Create filter
filters = np.zeros(shape= (7, 7, channel, 2), dtype= np.float32)
filters[:, 3, :, 0] = 1 # vertical line 
filters[3, :, :, 1] = 1  # horizontal line

outputs = tf.nn.conv2d(images, filters, strides= 1, padding= "SAME") # VALID skips some pixels unattended
plt.imshow(outputs[0, :, :, 1], cmap= "gray") # 2nd feature map of 1st image
plt.show()
```

```python
plt.imshow(outputs[1, :, :, 0], cmap= "gray") # 1st feature map of 2nd image
plt.show()
```
